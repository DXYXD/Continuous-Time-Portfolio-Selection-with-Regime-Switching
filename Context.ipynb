## 1. Background

To better reflect random market envionment, some finacial market models with key parameters, such as the bank interest rate, stocks appreciation rates, and volatility rates, are modulated by some Markov processes.

A factor that dominates the movement of a stock is the trend of the market. To reflect the market trend, it is necessary to allow the key parameters to respond to the general market movements. 

One of such formulations is the **regime switching model**, where the market parameters depend on the market mode that switches among a finite number of states.

The market mode could reflect the state of the underlying economy, the general mood of investors in the market, and other economic factors. 

A switching model can be formulated mathematically as a stochastic differential equation whose coefficients are modulated by a continuous-time Markov chain. 

## 2. Objective

- In this work, the paper focus on a continuous-time mean-variance model modulated by a Markov chain representing the regime switching.

- Develop a continuous-time version of the Nobel-prize winning mean-variance portfolio selection model with regime switching and attempt to derive closed-form solutions for efficient portfolios and efficient frontier.
